# 模拟登陆

有一个表单提交到这个URL，表单的两个内容分别为uid和password，学号和密码，没有其他的隐藏信息，提交方式为POST。

好，现在我们首先构造以下代码来完成登录。看看会不会获取到登录之后的提示页面。

```python
import urllib
import urllib2
import cookielib
import re

#天津大学绩点计算
class TJU():
    
    def __init__(self):
        self.loginURL = 'http://e.tju.edu.cn/Main/logon.do'
        #创建cookie对象
        self.cookies = cookielib.CookieJar()
        #创建提交的表单数据
        self.postdata = urllib.urlencode({
                'uid':'3011203252',
                'password':'a433991100'
            })
        #构建包含cookie的opener
        self.opener = urllib2.build_opener(urllib2.HTTPCookieProcessor(self.cookies))
    
    def getPage(self):
        #将表单数据封装成data构建成request
        request = urllib2.Request(
            url = self.loginURL,
            data = self.postdata        
        )
        #利用包含cookie的opener打开request
        result = self.opener.open(request)
        print result.read().decode('gbk')
        
```

```python
#test
tju = TJU()
tju.getPage()
```

```python
import urllib
import urllib2
import cookielib
import re

#天津大学绩点计算
class TJU():
    
    def __init__(self):
        #登陆账号地址
        self.loginURL = 'http://e.tju.edu.cn/Main/logon.do'
        self.semesterNum = raw_input('请输入学期号码') 
        #创建cookie对象
        self.cookies = cookielib.CookieJar()
        #创建提交的表单数据
        self.postdata = urllib.urlencode({
                'uid':'3011203252',
                'password':'a433991100'
            })
        #构建包含cookie的opener
        self.opener = urllib2.build_opener(urllib2.HTTPCookieProcessor(self.cookies))
    
    def getPage(self):
        #将表单数据封装成data构建成request
        request = urllib2.Request(
            url = self.loginURL,
            data = self.postdata        
        )
        #利用包含cookie的opener打开request
        result = self.opener.open(request)
        print result.read().decode('gbk')
     
    #输入学期名号，拿到对应成绩
    def gradePage(self):
        #执行之前请确保执行cookie确保登陆成功
        gradeURL = 'http://e.tju.edu.cn/Education/stuachv.do?todo=display&term=' + self.semesterNum
        #将表单数据封装成data构建成request
        request = urllib2.Request(
            url = self.loginURL,
            data = self.postdata        
        )
        #利用包含cookie的opener打开request执行登陆操作
        result = self.opener.open(request)
        #切换到gradePage
        result = self.opener.open(gradeURL)
        return result.read().decode('gbk')
        
    def gradeList(self,gradePage):
        pattern = re.compile('<font class=ContextText2>&nbsp;&nbsp;(.*?)</font></td>\s*.*<font .*?</font></td>\s*.*?<font class=ContextText2>.*?</font></td>\s*.*?Text2>(.*?)</font></td>\s*.*\s*(.*?)\s*</font>')
        return  re.findall(pattern,gradePage)
    
    def showgrade(self):
        print u"当前学期编号为:",self.semesterNum
        #获取该学期的表单数据
        gradepage = self.gradePage()
        gradelist = self.gradeList(gradepage)
        return gradelist
        print u'共有',len(gradelist),u'门科目'
        for item in gradelist:
            print '科目:',item[0],'   ','学分:',item[1],'   ','成绩:',item[2]
        
    def txtout(self):
        gradelist = self.showgrade()
        print gradelist
        try:
            f = open('grade.txt','w+')
            print >> f, "当前学期编号为",11121
            print >> f, '共有',len(gradelist),'门科目'
            for item in gradelist:
                print >> f, '科目:'+ item[0].encode('gbk'),'   ','学分:',item[1],'   ','成绩:',item[2]    
        except IOError,e:
            print "写入异常，原因",e.message
        finally:
            print "写入任务完成"
```
